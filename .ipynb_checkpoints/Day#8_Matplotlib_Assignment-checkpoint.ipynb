```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
plt.plot([1,2,3,4]);
```

```python
plt.plot(pd.DataFrame([9,5,3,7]));
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie(np.arange(1,5), labels=["CS", "IT", "IS", "MIS"], explode=[0.2,0.2,0.2,0.2]);
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
x = np.arange(500)
y = np.arange(500)
plt.scatter(x,y, alpha=0.4, s = 50, marker="o");
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(x,y, alpha=0.4, s = 50, marker="o");
plt.title("title",loc="left", c = "r")
```

## Q5: Create three line charts with Legends at left and colors

```python
plt.plot([1,2,3,4]);
plt.plot([5,6,7,8]);
plt.plot([9,10,11,12,13]);
plt.legend(["first", "second", "third"], shadow = True, loc = "upper left")
```

## Q6: Create subplots with three charts and figsize=20,10

```python
plt.figure(figsize=[20,10])
plt.suptitle('My_Subplots',fontsize = 20)
plt.subplot(2,2,1)
plt.plot([1,2,3,4]);
plt.subplot(2,2,2)
plt.pie(np.arange(1,5), labels=["CS", "IT", "IS", "MIS"], explode=[0.2,0.2,0.2,0.2]);
plt.subplot(2,2,3)
plt.bar(x, y)
plt.xlabel('Days')
plt.ylabel('Sales');
plt.grid()
```
